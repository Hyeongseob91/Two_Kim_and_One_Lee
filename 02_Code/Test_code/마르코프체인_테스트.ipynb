## 마르코프체인 테스트

```python
import os
import json
import random
from konlpy.tag import Okt
import csv

# 마르코프 체인 딕셔너리 만들기
def make_dic(words):
    tmp = ["@"]
    dic = {}
    for word in words:
        tmp.append(word)
        if len(tmp) < 3: continue
        if len(tmp) > 3: tmp = tmp[1:]
        set_word3(dic, tmp)
        if word == ".":
            tmp = ["@"]
            continue
    return dic

# 딕셔너리에 데이터 등록하기
def set_word3(dic, s3):
    w1, w2, w3 = s3
    if not w1 in dic: dic[w1] = {}
    if not w2 in dic[w1]: dic[w1][w2] = {}
    if not w3 in dic[w1][w2]: dic[w1][w2][w3] = 0
    dic[w1][w2][w3] += 1

# 문장 만들기
def make_sentence(dic):
    ret = []
    if not "@" in dic: return "no dic" 
    top = dic["@"]
    w1 = word_choice(top)
    w2 = word_choice(top[w1])
    ret.append(w1)
    ret.append(w2)
    while True:
        w3 = word_choice(dic[w1][w2])
        ret.append(w3)
        if w3 == ".": break
        w1, w2 = w2, w3
    return "".join(ret)

def word_choice(sel):
    keys = list(sel.keys())
    return random.choice(keys)

# 문장 읽어 들이기
track_file = "new_track.csv"   # 내가 수집한 가사 csv파일일
dict_file = "markov-track.json"

if not os.path.exists(dict_file):
    # CSV 파일 읽어 들이기
    text = ""
    with open(track_file, 'r', encoding='utf-8') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            text += row[3] + " "    # 내가 만든 데이터프레임에서 가사가 있는 열이 3번째라 숫자 3을 넣음.
    
    text = text.replace("…", "")
    
    # Okt를 사용한 형태소 분석
    okt = Okt()
    malist = okt.pos(text, norm=True)
    words = []
    for word in malist:
        if not word[1] in ["Punctuation"]:
            words.append(word[0])
        if word[0] == ".":
            words.append(word[0])
    
    # 딕셔너리 생성
    dic = make_dic(words)
    json.dump(dic, open(dict_file,"w", encoding="utf-8"))
else:
    dic = json.load(open(dict_file,"r"))

# 문장 만들기
for i in range(3):
    s = make_sentence(dic)
    print(s)
    print("---")
```
