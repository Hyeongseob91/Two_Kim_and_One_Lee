```python
import requests
from bs4 import BeautifulSoup

# 벅스뮤직 차트 URL
url = "https://music.bugs.co.kr/chart"

# HTTP 요청으로 페이지 가져오기
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# 곡 정보 수집 함수
def collect_song_info():
    songs = []

    # 곡 리스트 가져오기
    rows = soup.select(".list.trackList.byChart tbody tr")

    for row in rows:
        try:
            # 제목 수집
            title = row.select_one("th .title").get_text(strip=True)

            # 가수 수집
            artist = row.select_one("td .artist a").get_text(strip=True)

            # 발매일 및 상세 정보 수집
            release_date = row.select_one("td:nth-child(5)").get_text(strip=True)  # 발매일 정보 위치 수정 필요

            # 가사 링크가 있는 경우 가져오기
            lyrics_link = row.select_one("td .btnLyrics")
            if lyrics_link:
                lyrics_url = "https://music.bugs.co.kr" + lyrics_link["href"]
                lyrics_response = requests.get(lyrics_url)
                lyrics_soup = BeautifulSoup(lyrics_response.text, 'html.parser')
                lyrics = lyrics_soup.select_one("#lyrics p").get_text(strip=True)
            else:
                lyrics = "가사 정보 없음"

            # 좋아요 수 수집
            likes = row.select_one("td .like .count").get_text(strip=True).replace(',', '')

            # 곡 정보 저장
            songs.append({
                "title": title,
                "artist": artist,
                "release_date": release_date,
                "lyrics": lyrics,
                "likes": likes
            })
        except Exception as e:
            print(f"Error while processing a song: {e}")

    return songs

# 곡 정보 가져오기
song_data = collect_song_info()

# 결과 출력
for song in song_data:
    print(song)
```

```python
# 곡 정보 수집 함수작성
def collect_song_info():
    songs_data = dict()

    # 1. 노래 title 수집하기
    title_table = driver.find_element(By.CSS_SELECTOR, ".list.trackList.byChart")
    title_list = title_table.find_elements(By.CSS_SELECTOR, "tbody>tr")
    titles = []
    for title_component in title_list:
        titles.append(title_component.find_elements(By.CSS_SELECTOR, "th")[0].text)

    # 2. 노래 artist 수집하기
    artist_table = driver.find_element(By.CSS_SELECTOR, ".list.trackList.byChart")
    artist_list = artist_table.find_elements(By.CSS_SELECTOR, "tbody>tr")
    artist = []
    for artist_component in artist_list:
        artist.append(artist_component.find_elements(By.CSS_SELECTOR, "td>p>a")[0].text)

    # 3. 노래 lyrics 수집하기
    lyrics_table = driver.find_element(By.CSS_SELECTOR, ".list.trackList.byChart")
    lyrics_list = lyrics_table.find_elements(By.CSS_SELECTOR, "tbody>tr")
    lyrics_box = driver.find_element(By.PARTIAL_LINK_TEXT, '곡정보')
    lyrics_box.click()
    time.sleep(2)
    lyrics = []
    for lyrics_component in lyrics_list:
        lyrics.append(lyrics_component.find_elements(By.CSS_SELECTOR, "lyricsContainer>xmp")[0].text)

    # 4. 노래 release 수집하기
    release_table = driver.find_element(By.CSS_SELECTOR, ".list.trackList.byChart")
    release_list = release_table.find_elements(By.CSS_SELECTOR, "tbody>tr")
    try :
        release_box = driver.find_element(By.PARTIAL_LINK_TEXT, '곡정보')
        release_box.click()
        time.sleep(2)
    except :
        print(f"버튼 클릭 실패: {e}")
    release = []
    for release_component in release_list:
        release.append(release_component.find_elements(By.CSS_SELECTOR, "td>p>a")[0].text)

    # 5. 노래 likes 수집하기
    like_table = driver.find_element(By.CSS_SELECTOR, ".list.trackList.byChart")
    like_list = like_table.find_elements(By.CSS_SELECTOR, "tbody>tr")
    like = []
    for like_component in like_list:
        like.append(like_component.find_elements(By.CSS_SELECTOR, "td>p>a")[0].text)

    # 6. 정보 담기
    songs_data={
        "title": titles,
        "artist": artist,
        "lyrics": lyrics,
        "release": release,
        "likes": like
    }

    return songs_data
```
