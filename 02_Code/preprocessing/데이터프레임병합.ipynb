### 데이터프레임 병합

```python
import pandas as pd
import numpy as np
import re
```

```python
# CSV 파일 경로 리스트
file_paths = [
    '01_국내_랩힙합_data.csv',  # 첫 번째 CSV 파일 경로
    '02_국내_아이돌_data.csv',  # 두 번째 CSV 파일 경로
    '03_국내_알앤비_data.csv',  # 세 번째 CSV 파일 경로
    '04_국내_성인가요_data.csv',  # 네 번째 CSV 파일 경로
    '05_해외_팝_data.csv',  # 다섯 번째 CSV 파일 경로
    '06_해외_랩힙합_data.csv',   # 여섯 번째 CSV 파일 경로
    '07_해외_락_data.csv',
    'new_indie.csv'
]

# 빈 리스트 생성
dataframes = []

# 각 CSV 파일을 읽어서 리스트에 추가
for file in file_paths:
    df = pd.read_csv(file)  # CSV 파일 읽기
    dataframes.append(df)    # 리스트에 데이터프레임 추가

# 리스트에 있는 데이터프레임을 하나로 병합 (컬럼명이 다르더라도 병합)
merged_df = pd.concat(dataframes, ignore_index=True, join='outer')

# 결과 확인
merged_df

# # 병합된 데이터프레임을 새로운 CSV 파일로 저장 (선택 사항)
# merged_df.to_csv('merged_output.csv', index=False)
```

```python
merged_df.loc[merged_df['lyrics'].isna(), 'lyrics'] = merged_df['가사']
```

```python
merged_df.loc[merged_df['like_cnt'].isna(), 'like_cnt'] = merged_df['좋아요']
```

```python
merged_df.loc[merged_df['artist'].isna(), 'artist'] = merged_df['가수']
```

```python
merged_df.loc[merged_df['title'].isna(), 'title'] = merged_df['제목']
```

```python
df = merged_df.drop(["Unnamed: 0","제목","가수","가사","좋아요"],axis=1)
df
```

```python
for i in range(len(df)):
    cleaned_text = re.sub(r'[^a-zA-Z0-9가-힣\s]', '', df.loc[i, 'lyrics'])  # 특수 문자 제거
    cleaned_text = cleaned_text.replace('\n', ' ').replace('\r', ' ')  # 줄 바꿈 문자 제거
    df.loc[i, 'lyrics'] = cleaned_text.strip()  # 앞뒤 공백 제거
```

```python
df
```

```python
df.to_csv('merge_track.csv', index=False, encoding='utf-8-sig')
```

## 최종 데이터프레임 병합

```python
# CSV 파일 경로 리스트
file_paths = [
    'music_data.csv',  # 첫 번째 CSV 파일 경로
    'melon_data.csv',  # 두 번째 CSV 파일 경로
    'melon_indie.csv',  # 세 번째 CSV 파일 경로
    ]

# 빈 리스트 생성
dataframes = []

# 각 CSV 파일을 읽어서 리스트에 추가
for file in file_paths:
    df = pd.read_csv(file)  # CSV 파일 읽기
    dataframes.append(df)    # 리스트에 데이터프레임 추가

# 리스트에 있는 데이터프레임을 하나로 병합 (컬럼명이 다르더라도 병합)
merged_df = pd.concat(dataframes, ignore_index=True, join='outer')

# 결과 확인
merged_df
```

```python
merged_df.loc[merged_df['lyrics'].isna(), 'lyrics'] = merged_df['lyric']
```

```python
merged_df.loc[merged_df['likes'].isna(), 'likes'] = merged_df['like_cnt']
```

```python
merged_df.loc[merged_df['likes'].isna(), 'likes'] = merged_df['like']
```

```python
df = merged_df.drop(['Unnamed: 0','rank','songid','genre','like_cnt','like','lyric'], axis=1)
```
